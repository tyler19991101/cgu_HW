```python
(xr,yr), (xt,yt)= keras.datasets.mnist.load_data()
```

```python

xr.shape, yr.shape, xt.shape, yt.shape
```

```python
xr1= xr.reshape(-1, 28*28) 
xt1= xt.reshape(-1, 28*28)
```

```python
aModel= keras.Sequential([
    keras.Input(28*28),
    keras.layers.Dense(200),    
    keras.layers.Dense(100),
    keras.layers.Dense(50),
    keras.layers.Dense(10)   
    ])

aModel.compile(
    loss=     'sparse_categorical_crossentropy',
    metrics= ['accuracy']
    )
aModel.summary()

keras.utils.plot_model(aModel, 
    show_shapes= True, 
    show_layer_activations= True)
```

```python
aModel.fit(xr1, yr, 
           epochs= 10, 
           batch_size= 50)

aModel.evaluate(xt1, yt)
```

```python
bModel= keras.Sequential([
    keras.Input(28*28),
    keras.layers.Dense(200,activation= 'relu'),    
    keras.layers.Dense(100,activation= 'softmax'),
    keras.layers.Dense(50,activation= 'relu'),
    keras.layers.Dense(10,activation= 'softmax')   
    ])

bModel.compile(
    loss=     'sparse_categorical_crossentropy',
    metrics= ['accuracy']
    )
bModel.fit(xr1, yr, 
           epochs= 20, 
           batch_size= 100)

bModel.evaluate(xt1, yt)
```
